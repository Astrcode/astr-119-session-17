```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_samples = 10000
n_bin = 100
sigma = 1.0
mu = 0.0
```

```python
x = np.random.normal(mu,sigma,n_samples)
print(x.shape)
print(x.min())
print(x.max())
print(np.mean(x))
print(np.median(x))
print(np.std(x))
```

```python
def gaussian(x,mu,sigma):
    return 1./(2.0*np.pi*sigma**2)**0.5 * np.exp(-0.5*((x-mu)/sigma)**2)
```

```python
fig = plt.figure(figsize=(7,7))
y_hist, x_hist, ignored = plt.hist(x, bins=n_bin, range=[-5,5], density=True)
xx = np.linspace(-5.0,5.0,1000)
plt.plot(xx,gaussian(xx,mu,sigma),color="red")
plt.ylim([0,0.5])
plt.xlim([-5,5])
plt.gca().set_aspect(20)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.show()
```
